```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ventas.csv")
df2 = pd.read_csv("Empleadoss.csv")
```

```python
df.head()
```

```python
df2.head()
```

```python
df.dtypes
```

```python
df = df.astype({'Area':'string','CodigoFamilia':'string'})
```

```python
df
```

```python
df.dropna()
```

```python
df.dtypes
```

```python
df = pd.merge(df, df2, on='Empleado')
```

```python
df
```

```python
df = df.rename(columns={'Empleado':'IdEmpleado','Nombre y Apellido':'Empleado'})
```

```python
df = df.astype({'IdEmpleado':'string'})
```

```python
df
```

```python
df.dtypes
```

```python
len(df)
```

```python
df['Familia'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.countplot(data=df,x='Familia',hue='Area').set(title="Ventas por servicio y área");
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.barplot(data=df,x='Sede',y='Ventas',estimator=np.sum,hue="Familia",palette='Set2').set(title='Ventas totales por Sede'); 
plt.legend(bbox_to_anchor=(1,.7)); 
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.barplot(data=df,x='Sede',y='Ventas',estimator=np.mean,hue="Familia",palette='Set2').set(title='Ventas promedio por Sede'); 
plt.legend(bbox_to_anchor=(1,.7)); #moving the legend box
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.countplot(data=df,x='Sede',hue="Familia").set(title='Numero de ventas por Sede');
```

```python
plt.figure()
sns.displot(data=df,x='Sede',bins=10,kde=True,rug=True).set(title='Ventas por sede')
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="Ventas",y="Cantidad",data=df,hue="Sede",palette="tab10",).set(title="Relación Ventas y Cantidad"); 
```

```python
sns.lineplot(x='Año',y='Ventas',data=df,hue='Sede',ci=None,markers=["o",".","<","v",">"],style='Familia').set(title="Ventas anuales por sede")
plt.legend(bbox_to_anchor=(1.1,.5))
plt.xticks(rotation=25);
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.lmplot(x="Ventas",y="Cantidad",data=df,palette="tab10",).set(title="Relación Ventas y Cantidad"); 
```
